```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("Desktop/canada_per_capita_income.csv")
df.head()
```

```python
dt = df.rename(columns = {'per capita income (US$)':'per_capita_income'})
dt.head()
```

```python
%matplotlib inline
plt.xlabel('Year')
plt.ylabel('per capita income(US$)')
plt.scatter(dt.year,dt.per_capita_income)
plt.plot(dt.year,dt.per_capita_income)
```

```python
reg = linear_model.LinearRegression()
reg.fit(dt[['year']],dt.per_capita_income)
```

```python
reg.predict([[2030]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
%matplotlib inline
plt.xlabel('Year')
plt.ylabel('per capita income(US$)')
plt.scatter(dt.year,dt.per_capita_income)
plt.plot(dt.year,reg.predict(dt[['year']]))
```
