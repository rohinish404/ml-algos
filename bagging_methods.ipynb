```python
import pandas as pd
df = pd.read_csv('Desktop/csv_files/heart.csv')
df
```

```python
df1 = df[df.Cholesterol<=(df.Cholesterol.mean()+3*df.Cholesterol.std())]
df1.shape
```

```python
df2 = df1[df1.Oldpeak<=(df1.Oldpeak.mean()+3*df1.Oldpeak.std())]
df2.shape
```

```python
df3 = df2[df2.RestingBP<=(df2.RestingBP.mean()+3*df2.RestingBP.std())]
df3
```

```python
df4 = df3.copy()
df4.ExerciseAngina.replace(
    {
        'N': 0,
        'Y': 1
    },
    inplace=True)

df4.ST_Slope.replace(
    {
        'Down': 1,
        'Flat': 2,
        'Up': 3
    },
    inplace=True
)

df4.RestingECG.replace(
    {
        'Normal': 1,
        'ST': 2,
        'LVH': 3
    },
    
    inplace=True)
df4.Sex.replace(
    {
        'M': 0,
        'F': 1
    },
    
    inplace=True)
df4
dummies = pd.get_dummies(df4.ChestPainType)
df5 = pd.concat([df4,dummies],axis='columns')
df5
```

```python
df5.drop('ChestPainType',axis='columns', inplace=True)
df5
```

```python
X = df5.drop('HeartDisease',axis='columns')
Y = df5.HeartDisease
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X_scaled,Y,test_size=0.2)
```

```python
from sklearn import svm
from sklearn.model_selection import cross_val_score
scores1 = cross_val_score(svm.SVC(),X_scaled,Y,cv=5)
scores1
```

```python
scores1.mean()
```

```python
from sklearn.ensemble import BaggingClassifier
bag_model = BaggingClassifier(
    base_estimator = svm.SVC(),
    n_estimators=100,
    max_samples = 0.8,
    oob_score=True,
    random_state=0
)
bag_model.fit(x_train,y_train)
cross_val_score(bag_model,X_scaled,Y).mean()
```

```python
from sklearn import tree
scores2 = cross_val_score(tree.DecisionTreeClassifier(),X_scaled,Y)
```

```python
scores2.mean()
```

```python
bag_model2 = BaggingClassifier(
    base_estimator = tree.DecisionTreeClassifier(),
    n_estimators=50,
    max_samples = 0.8,
    oob_score=True,
    random_state=0
)
bag_model2.fit(x_train,y_train)
cross_val_score(bag_model2,X_scaled,Y).mean()
```

```python
from sklearn.ensemble import RandomForestClassifier

scores3 = cross_val_score(RandomForestClassifier(), X_scaled, Y, cv=5)
scores3.mean()
```
